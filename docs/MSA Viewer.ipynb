# BioJS MSA Viewer Example

```julia
fastainput = "https://raw.githubusercontent.com/diegozea/MIToS.jl/master/test/data/Gaoetal2011.fasta" 

display("text/html",
string("""
    <script src="https://s3.eu-central-1.amazonaws.com/cdn.bio.sh/msa/latest/msa.min.gz.js"></script>
  
    <script type="text/javascript">

    msa.io.fasta.read("$fastainput", function(err, seqs){

    var opts = {
        el: document.getElementById("yourDiv"),
        vis: {
            conserv: false,
            overviewbox: false,
            seqlogo:true
        },
    seqs: seqs,
    menu: "small",
    bootstrapMenu: false // IJulia
    };

    var m = new msa(opts).render();
    });

    </script>

    <div id="yourDiv">Run this IJulia cell with Internet connection to see the MSA.</div>
"""))
```
